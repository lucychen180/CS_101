```python
from os import listdir
import numpy as np
from numpy import asarray
from numpy import vstack
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
from numpy import savez_compressed
import ehtim as eh

# visibilities
INPUT_SIZE = 1030
# pixels in image
OUTPUT_SIZE = 784

IM = eh.image.load_txt('../models/avery_sgra_eofn.txt');
IM = IM.resample_square(28)

def displayImage(arr):
    """Displays image given a flattened array of pixels."""
    IM.imvec = arr
    IM.display()
```

```python
def getData(in_path, out_path, start, end):
    """Reads in visibilities and image from [start, end)."""
    n = end - start
    x, y = np.empty([n, INPUT_SIZE]), np.empty([n, OUTPUT_SIZE])
    for i in range(n):
        obs = eh.obsdata.load_uvfits((in_path + '/obs{0}.fits').format(start + i))
        im = eh.image.load_fits((out_path + '/img{0}.fits').format(start + i))
        x[i] = obs.data['vis']
        y[i] = im.imvec
    return x, y

# load dataset A
dataA1 = getData('../data/trainA', '../data/trainB', 0, 17000)
dataAB = getData('../data/testA', '../data/testB', 45000, 45100)
data = vstack((dataA1, dataAB))
print('Loaded data: ', data.shape)
# save as compressed numpy array
filename = 'obs2imvec.npz'
savez_compressed(filename, data)
print('Saved dataset: ', filename)

```
